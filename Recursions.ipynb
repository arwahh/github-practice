```python
def print_numbers(n):
    if n == 0:
        return
    print_numbers(n - 1)
    print(n)

print_numbers(5)
```

```python
def factorial(n):
    if n == 1:
        return 1
    return n * factorial(n - 1)

print(factorial(5))
```

```python
def sum_n(n):
    if n == 1:
        return 1
    return n + sum_n(n - 1)

print(sum_n(5))
```

```python
def print_digits(n):
    if n < 10:
        print(n)
        return
    print_digits(n // 10)
    print(n % 10)

print_digits(1234)
```

```python
def count_digits(n):
    if n == 0:
        return 0
    return 1 + count_digits(n // 10)

print(count_digits(54321))
```

```python
def fib(n):
    if n <= 1:
        return n
    return fib(n-1) + fib(n-2)

print(fib(6))
```

```python
def reverse(s):
    if len(s) == 0:
        return s
    return reverse(s[1:]) + s[0]

print(reverse("python"))
```

```python
def power(a, n):
    if n == 0:
        return 1
    return a * power(a, n - 1)

print(power(2, 3))
```

```python
def find_min(lst, n):
    if n == 1:
        return lst[0]
    return min(lst[n-1], find_min(lst, n-1))

print(find_min([5, 3, 8, 1, 9], 5))
```

```python
def sum_list(lst, n):
    if n == 0:
        return 0
    return lst[n-1] + sum_list(lst, n-1)

print(sum_list([1, 2, 3, 4, 5], 5))
```
