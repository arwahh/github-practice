```python
N = 10
total = 0

for i in range(1, N + 1):
    total += i

print("Sum =", total)
```

```python
nums = [3, 6, 2, 9, 12, 15, 18]

count = 0
for n in nums:
    if n % 2 == 0:
        count += 1

print("Even count =", count)
```

```python
text = "hello"
rev = ""

for ch in text:
    rev = ch + rev

print(rev)
```

```python
num = 7

for i in range(1, 11):
    print(num, "x", i, "=", num * i)
```

```python
nums = [10, 5, 22, 7, 15]

largest = nums[0]

for n in nums:
    if n > largest:
        largest = n

print("Largest =", largest)
```

```python
num = 13
is_prime = True

for i in range(2, num):
    if num % i == 0:
        is_prime = False
        break

if is_prime:
    print(num, "is Prime")
else:
    print(num, "is Not Prime")
```

```python
n1, n2 = 0, 1
print(n1, n2, end=" ")

for i in range(8):
    nxt = n1 + n2
    print(nxt, end=" ")
    n1 = n2
    n2 = nxt
```
