```python
students = {
    "Arwah": 85,
    "Seher": 92,
    "Falak": 78
}

print("Keys:", students.keys())
print("Values:", students.values())
```

```python
students["Seher"] = 90
print(students)
```

```python
students["Falak"] = 95
print(students)
```

```python
if "Falak" in students:
    print("Falak exists")
else:
    print("Not found")
```

```python
students.pop("Arwah")
print(students)
```

```python
for name, marks in students.items():
    print(name, ":", marks)
```

```python
topper = max(students, key=students.get)
print("Topper:", topper, students[topper])
```

```python
keys = ["name", "age", "city"]
values = ["Arooj", 22, "Delhi"]

d = dict(zip(keys, values))
print(d)
```

```python
word = "apple"
freq = {}

for ch in word:
    freq[ch] = freq.get(ch, 0) + 1

print(freq)
```
