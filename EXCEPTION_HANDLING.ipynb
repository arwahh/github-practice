```python
try:
    a = int(input("Enter numerator: "))
    b = int(input("Enter denominator: "))

    result = a / b

except ValueError:
    print("Error: Please enter valid integers only.")
except ZeroDivisionError:
    print("Error: Cannot divide by zero.")
else:
    # Runs only if no exception
    print("Result:", result)
finally:
    # Always runs
    print("Program finished.")
```

```python
nums = [10, 20, 30, 40, 50]

try:
    idx = int(input("Enter index (0 to 4): "))
    value = nums[idx]

except ValueError:
    print("Error: Index must be an integer.")
except IndexError:
    print("Error: Index out of range. Valid indexes are 0 to 4.")
else:
    print(f"Value at index {idx} is {value}")
```

```python
data = ["10", "20", "abc", "40", "5.5", "60"]
clean_numbers = []

for item in data:
    try:
        num = int(item)
    except ValueError:
        print(f"Skipping invalid value: {item}")
    else:
        clean_numbers.append(num)

print("Valid integers:", clean_numbers)
```

```python
def get_age():
    age_str = input("Enter your age: ")
    age = int(age_str)  # may raise ValueError

    if age < 0 or age > 120:
        raise ValueError("Age must be between 0 and 120.")

    return age


try:
    age = get_age()
except ValueError as e:
    print("Invalid age:", e)
else:
    print("Your age is", age)
```

```python
filename = input("Enter file name: ")

try:
    f = open(filename, "r")
    content = f.read()
except FileNotFoundError:
    print("Error: File not found.")
except Exception as e:
    print("Some other error occurred:", e)
else:
    print("File content:")
    print(content)
finally:
    print("Closing program...")
```

```python
class InsufficientBalanceError(Exception):
    """Custom exception raised when wallet balance is not enough."""
    pass


class Wallet:
    def __init__(self, balance=0):
        self.balance = balance

    def spend(self, amount):
        if amount > self.balance:
            raise InsufficientBalanceError("Not enough balance!")
        self.balance -= amount
        print(f"Remaining balance: {self.balance}")


# Using the Wallet
wallet = Wallet(500)

try:
    wallet.spend(700)
except InsufficientBalanceError as e:
    print("Transaction failed:", e)
else:
    print("Transaction successful.")
```
