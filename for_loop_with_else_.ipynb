```python
n = 29

for i in range(2, n):
    if n % i == 0:
        print(f"{n} is NOT a prime number.")
        break
else:
    print(f"{n} is a PRIME number.")
```

```python
nums = [10, 20, 30, 40, 50]
x = 40

for n in nums:
    if n == x:
        print("Found:", x)
        break
else:
    print("Not found:", x)
```

```python
text = "rhythm"

for ch in text:
    if ch.lower() in "aeiou":
        print("Vowel found:", ch)
        break
else:
    print("No vowel in the string")
```

```python
nums = [4, 8, 15, 16, 23, 42]

for n in nums:
    if n < 0:
        print("Negative number found:", n)
        break
else:
    print("No negative numbers in list")
```

```python
password = "Arwa@123"

for ch in password:
    if ch == " ":
        print("Invalid password â€“ contains space.")
        break
else:
    print("Password is Valid.")
```

```python
nums = [2, 4, 6, 8, 9]

for n in nums:
    if n % 2 != 0:
        print("Found an odd number:", n)
        break
else:
    print("All numbers are even")
```

```python
s = "programming"

for ch in s:
    if ch == "z":
        print("Character z found!")
        break
else:
    print("Character z NOT found.")
```
