```python
text = "   python is amazing. PYTHON is easy to learn! python provides many LIBRARIES.   "

# 1. Strip spaces
cleaned = text.strip()

# 2. Convert to lowercase
lowered = cleaned.lower()

# 3. Count "python"
count_python = lowered.count("python")

# 4. Replace a word
replaced = lowered.replace("amazing", "powerful")

# 5. Split into sentences using .replace() trick
sentences = replaced.replace("!", ".").split(".")

# Remove empty strings
sentences = [s.strip() for s in sentences if s.strip()]

# 6. Convert each sentence to title case
title_sentences = [s.title() for s in sentences]

# 7. Join back with |
final_text = " | ".join(title_sentences)

print("Count:", count_python)
print("Result:", final_text)
```

```python
s = "Welcome-To-The-World-Of-PythonProgramming123"

# 1. Count alphabets
alphabets = sum(c.isalpha() for c in s)

# 2. Count digits
digits = sum(c.isdigit() for c in s)

# 3. Check ending
ends_with_123 = s.endswith("123")

# 4. Split on '-'
parts = s.split("-")

# 5. Longest substring length
longest = max(parts, key=len)

# 6. Join with spaces
joined = " ".join(parts)

# 7. Check if title case
is_title = joined.istitle()

print("Alphabets:", alphabets)
print("Digits:", digits)
print("Ends with 123:", ends_with_123)
print("Longest part:", longest)
print("Joined:", joined)
print("Is Title:", is_title)
```

```python
name = "   Arwa    Fatima   "

# 1. Clean + lowercase
clean = name.strip().lower()

# 2. Split
parts = clean.split()

first = parts[0]
last = parts[-1]

# 3. Create username
user = first[:3] + last[-3:]

# 4. Ensure no spaces
user = user.replace(" ", "_")

# 5. Ensure length >= 6
if len(user) < 6:
    user = user.ljust(6, "x")

print("Generated Username:", user)
```

```python
password = "Password@123!!"

# 1. Lowercase
p = password.lower()

# 2. Keep only alphabets, digits, and '@'
cleaned = ''.join(ch for ch in p if ch.isalnum() or ch == '@')

# 3. Contains digits?
has_digits = any(ch.isdigit() for ch in cleaned)

# 4. Alphanumeric?
is_alnum = cleaned.isalnum()

# 5. Count vowels
vowels = sum(ch in 'aeiou' for ch in cleaned)

# 6. Contains 'pass'?
contains_pass = "pass" in cleaned

print("Cleaned:", cleaned)
print("Has Digits:", has_digits)
print("Is Alphanumeric:", is_alnum)
print("Vowel Count:", vowels)
print("Contains 'pass':", contains_pass)
```

```python
text = "   Python is GREAT. PYTHON is popular; Python is versatile!   "

# 1. Strip spaces
text = text.strip()

# 2. Lowercase
text_lower = text.lower()

# 3. Count 'python'
count_python = text_lower.count("python")

# 4. Replace 'great' with 'powerful'
text_lower = text_lower.replace("great", "powerful")

# 5. Normalize punctuation
for ch in [";", "!"]:
    text_lower = text_lower.replace(ch, ".")

# 6. Split into sentences
sentences = text_lower.split(".")

# 7. Remove empty
sentences = [s.strip() for s in sentences if s.strip()]

print("Python count:", count_python)
print("Processed sentences:")
for s in sentences:
    print(s.title())
```

```python
name = "   Arwa   seher   falak   "

# Clean and lowercase
clean = name.strip().lower()

# Split into parts
parts = clean.split()
first = parts[0]
last = parts[-1]

# Form username
username = first[:3] + last[-3:]

# Replace spaces just in case
username = username.replace(" ", "_")

# Ensure length >= 6
if len(username) < 6:
    username = username.ljust(6, "x")

print("Generated username:", username)
```

```python
email = "   Example.User123@Gmail.com   "

# 1 & 2
email = email.strip().lower()

# 3
if email.count("@") == 1:
    username, domain = email.split("@")

    # 4 & 5
    is_gmail = domain.endswith("gmail.com")

    # 6 Mask username
    if len(username) > 2:
        masked_user = username[:2] + "*" * (len(username) - 2)
    else:
        masked_user = username

    masked_email = masked_user + "@" + domain

    print("Original:", email)
    print("Is Gmail:", is_gmail)
    print("Masked:", masked_email)
else:
    print("Invalid email format")
```
