```python
for i in range(1, 11):
    if i == 7:
        break
    print(i)
```

```python
for i in range(1, 8):
    if i == 5:
        continue
    print(i)
```

```python
while True:
    word = input("Enter something: ")
    if word == "stop":
        break
    print("You entered:", word)
```

```python
for num in range(1, 11):
    if num % 2 != 0:   # odd â†’ skip
        continue
    print(num)
```

```python
nums = [5, 8, 3, -1, 6]

for n in nums:
    if n < 0:
        break
    print(n)
```

```python
names = ["seher", "Zoha", "Arwa", "falak"]

for n in names:
    if n.startswith("A"):
        continue
    print(n)
```

```python
for i in range(1, 4):
    for j in range(1, 5):
        if i == 2 and j == 3:
            break
        print(i, j)
```

```python
for i in range(1, 4):
    for j in range(1, 4):
        if j == 2:
            continue
        print(i, j)
```
