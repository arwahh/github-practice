```python
fruits = ["apple", "banana", "mango", "orange"]
print(fruits[0])   # first item
print(fruits[-1])  # last item
```

```python
fruits = ["apple", "banana"]
fruits.append("mango")
fruits.insert(1, "orange")
print(fruits)
```

```python
fruits = ["apple", "banana", "mango"]
fruits.remove("banana")
fruits.pop()
print(fruits)
```
