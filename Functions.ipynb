```python
def greet():
    print("Hello Python")

greet()
```

```python
def hello(name):
    print("Hello", name)

hello("Arwa")
```

```python
def square(num):
    return num * num

print(square(5))
```

```python
def greet(name="User"):
    return "Hello " + name

print(greet())
print(greet("Arwa"))
```

```python
def add(a, b):
    return a + b

print(add(10, 20))
```

```python
def even_odd(num):
    if num % 2 == 0:
        return "Even"
    else:
        return "Odd"

print(even_odd(7))
```

```python
def total(*nums):
    return sum(nums)

print(total(1, 2, 3, 4))
```

```python
def factorial(n):
    if n == 1:
        return 1
    return n * factorial(n - 1)

print(factorial(5))
```

```python
def person(**details):
    return details

print(person(name="Arwa", age=20))
```

```python
multiply = lambda a, b: a * b
print(multiply(3, 4))
```
