```python
fruits = ["apple", "banana", "mango", "kiwi"]

for index, fruit in enumerate(fruits):
    print(index, "→", fruit)
```

```python
fruits = ["apple", "banana", "mango", "kiwi"]

for i, fruit in enumerate(fruits, start=1):
    print(i, ".", fruit)
```

```python
nums = [10, 20, 30, 40, 50]

for i, val in enumerate(nums):
    if val == 30:
        print("Index of 30 is:", i)
        break
```

```python
nums = [12, 7, 18, 5, 22, 9]

for i, n in enumerate(nums):
    if n % 2 == 0:
        print(f"Even number {n} found at index {i}")
```

```python
text = "Python"

for i, ch in enumerate(text):
    print(i, "→", ch)
```

```python
items = ["A", "B", "C", "D", "E"]

for i, val in enumerate(items):
    if i % 2 == 1:
        items[i] = "*"

print(items)
```

```python
colors = ["red", "green", "blue"]

result = {i: color for i, color in enumerate(colors)}
print(result)
```

```python
text = "Python programming is fun"
words = text.split()

for i, word in enumerate(words, start=1):
    print(f"Word {i}: {word}")
```

```python
s = "rhythmgeek"

for i, ch in enumerate(s):
    if ch in "aeiou":
        print("First vowel:", ch)
        print("Index:", i)
        break
```

```python
names = ["Arwa", "Fatima", "Zoya"]
marks = [88, 92, 79]

for i, (name, mark) in enumerate(zip(names, marks), start=1):
    print(f"{i}. {name} → {mark}")
```
